```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('E:datamining-master/dataset_soal No. 3.csv', usecols=['Usia','Kelahiran_ke-','Waktu_Kelahiran','Tekanan_darah','Kelainan_jantung','Caesarian'], 
                        sep=';',dtype={'Usia': 'int32', 'Kelahiran_ke-': 'int32', 'Waktu_Kelahiran': 'int32', 'Tekanan_darah': 'int32', 'Kelainan_jantung': 'int32', 'Caesarian': 'int32'}) 
```

```python
df
```

```python
import math 
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia'])-30)**2+
                         (float(df.iloc[i]['Kelahiran_ke-'])- 1)**2+
                         (float(df.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(df.iloc[i]['Tekanan_darah'])-1)**2))
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
y = df.sort_values('dis').head(12)
y
```

```python
z = y["Caesarian"]
z
```

```python
np.mean(z)
```
